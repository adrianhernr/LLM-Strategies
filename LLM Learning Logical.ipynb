# Tests. Language models. Code execution.

Language models. We are going to evaluate the code execution capabilities.

Test 0.0 For a given function, extract properties and data. Wrong solution.

````python
import os
import openai

# Load your API key
openai.api_key=""

response = openai.Completion.create(
    engine="text-davinci-003",
    prompt='What are the real-valued roots of the following polynomial $x^{3}+2(x^{2})-1$?',
    temperature=0.5,
    max_tokens= 1000,
)
print(response["choices"][0]["text"])
````

Test 0.1 For a given function, extract properties and data using code execution.

OpenAI provides models with a working Python interpreter in a sandboxed, firewalled execution environment, along with some ephemeral disk space. Code run by the interpreter plugin is evaluated in a persistent session that is alive for the duration of a chat conversation. Also provides support uploading files to the current conversation workspace and downloading the results of your work.

````python
import os
import openai

# Load your API key
openai.api_key=""

completion = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "You can write and execute Python code by enclosing it in triple backticks, \
    e.g. ```code goes here```. Use this to run the code and perform calculations."},
    {"role": "user", "content": "Write and run Python code to find and print all roots of the following real variable polynomial: x**4-2*(x**2)=0.\
    Please run the code and print the roots"}
  ]
)
````

````python
print(completion.choices[0].message["content"])
````

The code executed in the notebook.

````python
import sympy as sp

# Define the polynomial
x = sp.Symbol('x')
polynomial = x**4 - 2*(x**2)

# Find the roots
roots = sp.solve(polynomial, x)

# Print the roots
for root in roots:
    print(root)
````
