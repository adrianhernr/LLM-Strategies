# Learning Strategies. Objectives of the experiments

In our research we have seen that a specific capability of humans is to generate new knowledge through learning strategies. Current artificial intelligence and deep learning models are very static and once trained they do not generate new information.

In machine learning, we define a learning strategy as a sequence of actions to select some data, define a model with all its meta-parameters, implement it, generate knowledge and improve the model.

In these experiments we want to evaluate the ability of language models to generate learning strategies that generate new knowledge using external tools and data: Data, agents, code interpreters, etc. Some external resources are now implemented by chatgp 3.5 or 4, as the code interpreter.

The sequence of actions is as follows:

1.   Language model. Define a model to solve a problem. Specify the model code, metaparameters, data and the model working.
2.   Language model or external interpreter. Implement the model, generate input-output, etc.
3.   Language model. Generate conclussions of the model, generate input-output, etc.
4.   Language model. With the model (code, parameters, input-output data, conclusions) and data, how we can we complement this model.
5.   Iterate the same process.

The structure of the input prompt to the language model will normally be:

**Role: System.** Content: It depends on the phase of the strategy and may have additional input data. Input data: {Input data}.

In the initial phase, it explains what a learning strategy is: "A machine learning strategy is a set of activities to learn from the data and solve a problem: A set of training data and a model to apply it".

In the next phase, explain that we are solving a problem, we have a model and some data in the input and we want to complement it with all the input information. Input data: {code, parameters, input-output data, more training data, conclusions}

**Role: Assitant.** Content: It depends on the phase of the strategy and may have additional example data. Example data: {Example data}.

In the initial phase we give a example: E.g. We want to learn a model for predicting oil prices. \
Output: Data: World Bank, IMF. Code: Python code of a neural network with input variables world GDP, GDP forecast, supply growth, etc.

In the next phase we give a example on how to complement a model with all the input information. E.g. We can add new input variables. We can relate the model variables. We can generate input-output data and relate the information. We can look for new data and complement the model.   

**Role: User.** Content: It depends on the phase of the strategy and may have additional input data. In the initial phase we ask for a learning strategy. In the next phase we ask it to complement the model.  

The following aspects are evaluated:
1. Ability to define an adequate model for the problem with all the necessary inputs and meta-parameters. Is the concept of learning strategy correctly understood? Does the model incorporate the problem adequately? Have all the phases of the model, hyperparameters, external data, etc. been defined?
2. Ability to run the model. Execution problems. Errors in the code. Autonomous execution.
3. Ability to obtain new information using the model: generate input-output, generate conclusions, new relationships, etc.
4. Ability to, based on the model, complement it or define new models.


# Machine learning strategy. Example.

Test 1.0 Learning strategy is a sequence of steps to define a machine learning model, select data and implement the model to solve a problem. Example of a learning strategy. We want you to define a learning strategy (machine learning model code and data) example.

```python
response = openai.Completion.create(
    engine="text-davinci-003",
    prompt='We have defined a machine learning strategy as a set of activities \
    to learn from the data: A set of training data and a model to apply it.\
    \n\n Example:We want to learn a model for predicting oil prices. \
    Output: Data: World Bank, IMF. Code: Python code of a neural network with input variables world GDP, GDP forecast, supply growth, etc.\n\n \
    Can you give me another example of learning strategy providing the source of the training data and the source code in Python. We want you to write the code in Python?',
    temperature=0.5,
    max_tokens= 2000,
)
print(response["choices"][0]["text"])
```

# Strategy and model definition

Test 1.0 Learning strategy is a sequence of steps to define a machine learning model, select data and implement the model to solve a problem.  We want you to define a learning strategy (machine learning model code and data) to solve a problem. July 2023.

```python

completion = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "You are a machine learning engineer. The user will provide you with a problem and a requirements field and you have to define a machine learning strategy with those requirement. \
    A machine learning strategy is a set of instructions to learn from the data and solve a problem: A set of training data and a model to apply it.\
    Use the following step-by-step instructions to define the machine learning strategy:\
    1. Specify data collection with data sources.\
    2. Justify model selection and training strategies.\
    3. Write the Python code of the learning strategy. Write the detailed code for data loading, model definition and model training."},
    {"role": "user", "content":" Can you give me a learning strategy to predict stock prices? \
    Requirements: {Python library: PyTorch. Possible models: Any machine learning model but preferably neural networks}."}
  ]
)
```

```python
print(completion.choices[0].message.content)
```

Evaluation. 1.0 The language model (chatGPT 3.5) is able to define an adequate model for the problem with all the necessary inputs and meta-parameters. This model can create new information.

1.1 The concept of learning strategy is correctly understood. The model specifies a set of instructions to learn from the data and solve a problem. The model called this a machine learning strategy.

1.2 It is important to emphasize that all actions can be implemented automatically: The data can be obtained through a link or an API and
the model could be implemented by chatGPT using its sandbox or implemented externally.  

1.3 All the phases of the model (data loading, model definition, model training, model evaluation, etc.) hyperparameters, external data, etc. have been stated or defined and explained.
